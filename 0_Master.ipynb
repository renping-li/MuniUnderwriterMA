```python
# This function runs a Stata script
def rundofile(filename):

    import subprocess
    
    Main_Path = "/Users/renpingli/Dropbox/MunicipalBond/Codes_MuniMarketPower/"

    dofile = '"' + Main_Path + filename + '"'
    cmd = ["/Applications/Stata/StataMP.app/Contents/MacOS/StataMP", "do", dofile]
    subprocess.call(cmd) 

# This funciton runs a notebook
def runnotebook(filename):

    import subprocess
    import nbformat
    from nbconvert.preprocessors import ExecutePreprocessor

    Main_Path = "/Users/renpingli/Dropbox/MunicipalBond/Codes_MuniMarketPower"

    notebook = Main_Path + filename
    with open(notebook) as f:
        nb = nbformat.read(f, as_version=4)
    ep = ExecutePreprocessor(timeout=-1, kernel_name='python3')
    ep.preprocess(nb, {'metadata': {'path': Main_Path + "/"}})
    with open(notebook, 'w', encoding='utf-8') as f:
        nbformat.write(nb, f)
```
