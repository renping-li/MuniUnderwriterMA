# 2. Calculate Sim Scores

## 2.1 Represent Market Share of State X Year as Vectors

```python
# Create a three-layer (State X Year X Underwriter) dictionary of market share

GNI = GNI.reset_index(drop=True)
GNI = GNI[GNI['State']!="Puerto Rico"]
GNI = GNI[GNI['State']!="American Samoa"]
GNI = GNI[GNI['State']!="Foreign"]
GNI = GNI[GNI['State']!="Unknown"]
GNI = GNI[GNI['State'].astype(str)!="nan"]

GNI = GNI[~pd.isnull(GNI['PrincipalAmount'])]

AllManagers = []
for idx,row in GNI.iterrows():
    if row['All Managers']==None or str(row['All Managers'])=='nan':
        continue
    Managers = row['All Managers'].split('\n')
    Managers = [FUN_proc_name(manager) for manager in Managers]
    AllManagers = AllManagers+Managers
    if idx%100==0:
        AllManagers = list(set(AllManagers))

Years = list(GNI['IssueYear'].unique())
States = list(GNI['State'].unique())

nested_dict = {item1: {item2: {item3: {'CorpBondVolume':0,'CorpEquityVolume':0} 
    for item3 in AllManagers} for item2 in Years} for item1 in States}

# Record volume of each State X Year X Underwriter X Security Type
# If using "row['PrincipalAmount']/n_Managers", would be using dollar amount, which I do not for consistency with HHI in the main test

for idx,row in GNI.iterrows():
    if row['All Managers']==None or str(row['All Managers'])=='nan':
        continue
    Managers = row['All Managers'].split('\n')
    n_Managers = len(Managers)
    for manager in Managers:
        if row['IsEquity']==False:
            nested_dict[row['State']][row['IssueYear']][FUN_proc_name(manager)]['CorpBondVolume'] = \
                nested_dict[row['State']][row['IssueYear']][FUN_proc_name(manager)]['CorpBondVolume']+ \
                1/n_Managers
        if row['IsEquity']==True:
            nested_dict[row['State']][row['IssueYear']][FUN_proc_name(manager)]['CorpEquityVolume'] = \
                nested_dict[row['State']][row['IssueYear']][FUN_proc_name(manager)]['CorpEquityVolume']+ \
                1/n_Managers

Volumes = pd.DataFrame([(outerKey, innerKey, innermoreKey, innermostKey, value) 
                   for outerKey, innerDict in nested_dict.items() 
                   for innerKey, innermoreDict in innerDict.items() 
                   for innermoreKey, innermostDict in innermoreDict.items() 
                   for innermostKey, value in innermostDict.items()],
                  columns=['State', 'IssueYear', 'Manager', 'SecurityType','Value'])

# Number of deals

Count = GNI.groupby(['State','IssueYear','IsEquity']).agg({'PrincipalAmount':'count'})
Count = Count.reset_index()
Count['SecurityType'] = None
Count.loc[Count['IsEquity'],'SecurityType'] = 'CorpEquityVolume'
Count.loc[~Count['IsEquity'],'SecurityType'] = 'CorpBondVolume'
Count = Count.drop(columns=['IsEquity'])
Count = Count.rename(columns={'PrincipalAmount':'n_deals'})

# Calculate market share

Volumes_sum = Volumes.groupby(['State','IssueYear','SecurityType']).agg({'Value':sum})
Volumes_sum = Volumes_sum.rename(columns={'Value':'Value_sum'})
Volumes_sum = Volumes_sum.reset_index()
Volumes_sum.loc[Volumes_sum['Value_sum']<0.001,'Value_sum'] = None
Volumes = Volumes.merge(Volumes_sum,on=['State','IssueYear','SecurityType'])
Volumes['market_share'] = Volumes['Value']/Volumes['Value_sum']
Volumes = Volumes.merge(Count,on=['State','IssueYear','SecurityType'])

Volumes_gb = Volumes.groupby(['State','IssueYear','SecurityType'])

Volumes_Corp = Volumes
```

## 2.2 Sim score for bond

```python
%%time

# Calculate similarity for each pair of states

SimScoreBond = []
for stateA in States:
    for stateB in States:
        if stateA==stateB:
            continue
        for year in Years:
            # Skip if there is no deals
            if np.sum((Count['State']==stateA)&(Count['IssueYear']==year)&(Count['SecurityType']=='CorpBondVolume'))==0 or \
                np.sum((Count['State']==stateB)&(Count['IssueYear']==year)&(Count['SecurityType']=='CorpBondVolume'))==0:
                continue
            DataA = Volumes_gb.get_group((stateA,year,'CorpBondVolume'))
            DataB = Volumes_gb.get_group((stateB,year,'CorpBondVolume'))
            # Record similarity score in terms of corporate equity underwriter, corporate bond underwriter, 
            # along with volumes in number of deals and dollars
            DataA = DataA[['Manager','market_share','n_deals']].rename(columns={'market_share':'market_shareA','n_deals':'NDealsA'})
            DataB = DataB[['Manager','market_share','n_deals']].rename(columns={'market_share':'market_shareB','n_deals':'NDealsB'})
            Data = DataA.merge(DataB,on=['Manager'])
            Data = Data.reset_index(drop=True)
            Cosine = np.dot(Data['market_shareA'],Data['market_shareB'])/ \
                (np.linalg.norm(Data['market_shareA'])*np.linalg.norm(Data['market_shareB']))
            SimScoreBond = SimScoreBond+[{
                'StateA':stateA,
                'StateB':stateB,
                'Year':year,
                'Cosine':Cosine,
                'NDealsA':Data.at[0,'NDealsA'],
                'NDealsB':Data.at[0,'NDealsB'],
                }]
SimScoreBond = pd.DataFrame(SimScoreBond)
```

```python
# Calculate similarity for each pair of states

Volumes_gb = Volumes.groupby(['State','IssueYear'])

SimScoreMuni = []
for stateA in States:
    for stateB in States:
        if stateA==stateB:
            continue
        for year in Years:
            # Skip if there is no deals
            if np.sum((Count['State']==stateA)&(Count['IssueYear']==year))==0 or \
                np.sum((Count['State']==stateB)&(Count['IssueYear']==year))==0:
                continue
            DataA = Volumes_gb.get_group((stateA,year))
            DataB = Volumes_gb.get_group((stateB,year))
            # Record similarity score in terms of corporate equity underwriter, corporate bond underwriter, 
            # along with volumes in number of deals and dollars
            DataA = DataA[['Manager','market_share','n_deals']].rename(columns={'market_share':'market_shareA','n_deals':'NDealsA'})
            DataB = DataB[['Manager','market_share','n_deals']].rename(columns={'market_share':'market_shareB','n_deals':'NDealsB'})
            Data = DataA.merge(DataB,on=['Manager'])
            Data = Data.reset_index(drop=True)
            Cosine = np.dot(Data['market_shareA'],Data['market_shareB'])/ \
                (np.linalg.norm(Data['market_shareA'])*np.linalg.norm(Data['market_shareB']))
            SimScoreMuni = SimScoreMuni+[{
                'StateA':stateA,
                'StateB':stateB,
                'Year':year,
                'Cosine':Cosine,
                'NDealsA':Data.at[0,'NDealsA'],
                'NDealsB':Data.at[0,'NDealsB'],
                }]
SimScoreMuni = pd.DataFrame(SimScoreMuni)
SimScoreMuni = SimScoreMuni[(SimScoreMuni['NDealsA']>=5)&(SimScoreMuni['NDealsB']>=5)]
```
